```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000
#TODO

def get_value (row):
   if row['LIMIT_BAL'] <= 10000 :
      return 'A'
   if (row['LIMIT_BAL'] <= 100000) & (row['LIMIT_BAL'] > 10000) :
      return 'B'
   if (row['LIMIT_BAL'] <= 200000) & (row['LIMIT_BAL'] > 100000) :
      return 'C'
   if (row['LIMIT_BAL'] <= 400000) & (row['LIMIT_BAL'] > 200000) :
      return 'D'
   if (row['LIMIT_BAL'] <= 700000) & (row['LIMIT_BAL'] > 400000) :
      return 'E'
   if (row['LIMIT_BAL'] > 700000) :
      return 'F'
   return 'Other'

df["NEW_COL"] = df.apply (lambda row: get_value(row), axis=1)
```

```python
# (8) построить зависимость кредитного лимита от образования для каждого пола
# график необходимо сделать очень широким (на весь экран)
#TODO
import matplotlib.pyplot as plt

plt.title("Зависимость кредитного лимита от образования для каждого пола")
plt.xlabel("Образование")
plt.ylabel("Лимит")

plt.grid(True, linestyle='-', color='0.75')

#fig_size = plt.rcParams["figure.figsize"]

x = sorted(df["EDUCATION"].unique())

y_male = []
y_female = []

for education in x:
    y_male.append(df[(df['SEX'] == 1) & (df['EDUCATION'] == education)]['LIMIT_BAL'].mean())
    y_female.append(df[(df['SEX'] == 2) & (df['EDUCATION'] == education)]['LIMIT_BAL'].mean())
    
#plt.set_size_inches(forward=True)
     
plt.plot(x, y_male, color="blue", label="мужчины")
plt.plot(x, y_female, color="red", label="женщины")

plt.show()
```
